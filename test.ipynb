$$ f(x) = ax^2 + bx + c $$

$$ x = {-b \pm \sqrt{b^2-4ac} \over 2a} $$


```python
def f(a, b, c) -> tuple:
    x1 = (-b + ((b^2 -4*a*c)**(0.5)))/(2 * a)
    x2 = (-b - ((b^2 -4*a*c)**(0.5)))/(2 * a)
    return (x1, x2)
```

```python
print(f(1, 2, -4))
```
